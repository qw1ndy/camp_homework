```python
import numpy as np
import matplotlib.pyplot as plt
import random

np.random.seed(2023)
heights = np.random.randint(1,100,20)

print(heights)
```

```python
def find_deepest_lake(heights):
    n = len(heights)
    max_depth = 0
    deepest_lake_bounds = (0, 0)
    
    for i in range(n):
        for j in range(i + 2, n):
            min_value = min(heights[i], heights[j])
            inner_min = min(heights[i+1:j])
            depth = min_value - inner_min
            
            if depth > max_depth:
                max_depth = depth
                deepest_lake_bounds = (i, j)
    
    return max_depth, deepest_lake_bounds
```

```python
def plot_heights_with_lake(heights, lake_bounds):
    x = np.arange(len(heights))
    
    plt.figure(figsize=(12, 6))
    plt.plot(x, heights, marker='o', linestyle='-', color='black', label='Heights')
    
    if lake_bounds != (0, 0):
        i, j = lake_bounds
        plt.fill_between(x[i:j+1], heights[i:j+1], min(heights[i], heights[j]), color='red', alpha=0.4, label='Deepest Lake')
    
    plt.xlabel("Index")
    plt.ylabel("Height")
    plt.title("Visualization of Heights and Deepest Lake")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
deepest_depth, lake_bounds = find_deepest_lake(heights)
print(f"Deepest lake depth: {deepest_depth}")
print(f"Lake formed between indices: {lake_bounds}")

plot_heights_with_lake(heights, lake_bounds)
```
