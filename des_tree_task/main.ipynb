```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import plot_tree 
from xgboost import XGBClassifier
# import graphviz
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
cancer = load_breast_cancer()
X, y, labels, features = cancer.data, cancer.target, cancer.target_names, cancer.feature_names
print ('labels:',labels)
print ('features:', features)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
des_tree_clf = DecisionTreeClassifier(criterion="entropy")
des_tree_clf.fit(X_train_scaled, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(des_tree_clf.score (X_train_scaled, y_train)))
print("test accuracy= {:.3%}".format(des_tree_clf.score (X_test_scaled, y_test)))
# graph_viz = tree.export_graphviz(clf, out_file=None,feature_names=features, class_names=labels, filled = True)
# graph = graphviz.Source(graph_viz)
# graph.view(cleanup=True) # cleanup (bool) – Delete the source file after rendering.
plt.figure(figsize=(20, 10))
plot_tree(des_tree_clf, feature_names=features, class_names=labels, filled=True)
plt.show
```

```python
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
rand_forest_clf = RandomForestClassifier()
rand_forest_clf.fit(X_train_scaled, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(rand_forest_clf.score (X_train_scaled, y_train)))
print("test accuracy= {:.3%}".format(rand_forest_clf.score (X_test_scaled, y_test)))
```

```python
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
grad_boost_clf = GradientBoostingClassifier()
grad_boost_clf.fit(X_train_scaled, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(grad_boost_clf.score (X_train_scaled, y_train)))
print("test accuracy= {:.3%}".format(grad_boost_clf.score (X_test_scaled, y_test)))
```

```python
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
XG_boost_clf = XGBClassifier()
XG_boost_clf.fit(X_train_scaled, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(XG_boost_clf.score (X_train_scaled, y_train)))
print("test accuracy= {:.3%}".format(XG_boost_clf.score (X_test_scaled, y_test)))
```
